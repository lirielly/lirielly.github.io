### Instalando o R no Jupyter Notebook - Método 1: quando você não tem o R / RStudio instalado e quer usar um ambiente virtual para seu projeto em R


+ Abrir Anaconda Promt

+ Criar seu environment para o projeto


```
    conda create --name <nome_do_seu_env>
```

+ Ativar o seu environment

```
    conda activate <nome_do_seu_env>
```

+ Instalar as bibliotecas que você vai precisar para o seu projeto segue os comandos para instalar o R e o Jupyter Notebook

```
    conda install -c r r-essentials
    
    conda install -c anaconda jupyter
```

+ Abrir o jupyter notebook 
```
   jupyter notebook
```


___

### Instalando o R no Jupyter Notebook - Método 2: quando você já possui o R / RStudio instalado 


+ Abrir Anaconda Prompt

+ Navegar ate a pasta do R:
```
    cd C:\Program Files\R\R-3.5.2\bin\x64
```

+ Executar o R:
```
    R.exe
```

+ Instalar os seguintes pacotes:
```
    install.packages("devtools")
    install.packages('IRkernel')
    IRkernel::installspec()
```
